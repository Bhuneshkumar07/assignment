```python
import numpy as np

matrix = np.array([[10, 21, 32, 43],
                   [54, 65, 76, 87],
                   [98, 19, 20, 31],
                   [42, 53, 64, 75]])

print("Original Matrix:\n", matrix)

print("Diagonal Elements:", np.diag(matrix))

matrix[matrix % 2 == 0] = 0
print("After Replacing Even Numbers:\n", matrix)

print("Index of Maximum Value:", np.argmax(matrix))
```

```python
import pandas as pd

data = {
    "Name":["Aman","Riya","Nikhil","Kriti","Rohan",
            "Priya","Ankit","Sneha","Rahul","Neha"],
    "Age":[20,21,19,22,20,21,20,19,22,20],
    "Marks":[85,72,91,67,78,88,74,95,81,69],
    "City":["Jaipur","Delhi","Pilani","Mumbai","Ajmer",
            "Delhi","Jaipur","Udaipur","Pilani","Kota"]
}

df = pd.DataFrame(data)

print("First 5 Rows:")
print(df.head())

print("\nStatistics:")
print(df.describe())

print("\nMarks > 75:")
print(df[df["Marks"] > 75])

print("\nSorted by Marks:")
print(df.sort_values(by="Marks", ascending=False))
```

## ques6

```python
import pandas as pd
import numpy as np

data = {
    "Name":["Aman","Riya","Nikhil","Kriti"],
    "Marks":[85,np.nan,90,np.nan],
    "City":["Jaipur",None,"Delhi",None]
}

df = pd.DataFrame(data)

print("Missing Values:")
print(df.isnull())

mean_marks = df["Marks"].mean()
df["Marks"] = df["Marks"].fillna(mean_marks)

print("\nAfter Filling Marks:")
print(df)

df2 = df.dropna(subset=["City"])

print("\nAfter Dropping Missing City:")
print(df2)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    "StudentID":[101,102,103,104],
    "Name":["Aman","Riya","Nikhil","Kriti"],
    "Department":["CSE","ECE","IT","CSE"]
})

df2 = pd.DataFrame({
    "StudentID":[101,102,103,104],
    "Marks":[92,75,88,96],
    "Grade":["A","B","A","A"]
})

merged = pd.merge(df1, df2, on="StudentID")

print("Merged DataFrame:")
print(merged)

print("\nStudents with Grade A:")
print(merged[merged["Grade"] == "A"])
```

```python
import matplotlib.pyplot as plt

subjects = ["Math","Physics","Chemistry","Python","English"]
marks = [80,85,78,92,88]

plt.figure(figsize=(6,4))
plt.plot(subjects, marks, marker='o', label="Marks")
plt.title("Marks Trend Across Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)
plt.show()

students = ["A","B","C","D","E"]
scores = [78,85,90,72,88]

plt.figure(figsize=(6,4))
plt.bar(students, scores, label="Marks")
plt.title("Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Marks":[65,70,75,80,85,90,95,88,76,69],
    "Math":[78,85,90,67,88,92,75,80,84,79],
    "Science":[72,89,91,70,82,95,77,83,86,74],
    "English":[80,78,88,69,85,90,76,84,82,75]
})

df["Marks"].plot(kind="hist", title="Marks Distribution")
plt.show()

df["Marks"].plot(kind="box", title="Box Plot")
plt.show()

category = pd.Series([40,30,20,10],
                     index=["A","B","C","D"])

category.plot(kind="pie", autopct="%1.1f%%",
              title="Category Percentage")
plt.ylabel("")
plt.show()
```

## ques11

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(1)

df = pd.DataFrame({
    "Age": np.random.randint(22,50,50),
    "Salary": np.random.randint(25000,100000,50),
    "Department": np.random.choice(
        ["HR","IT","Finance"],50),
    "Experience": np.random.randint(1,20,50)
})

sns.scatterplot(data=df,
                x="Age",
                y="Salary",
                hue="Department")
plt.title("Age vs Salary")
plt.show()

sns.heatmap(df[["Age","Salary","Experience"]].corr(),
            annot=True)
plt.title("Correlation Heatmap")
plt.show()

sns.barplot(data=df,
            x="Department",
            y="Salary")
plt.title("Average Salary per Department")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,11)
y = x * 2

fig, ax = plt.subplots(2, 2, figsize=(10,8))

ax[0,0].plot(x, y)
ax[0,0].set_title("Line Plot")

ax[0,1].scatter(x, y)
ax[0,1].set_title("Scatter Plot")

ax[1,0].bar(x, y)
ax[1,0].set_title("Bar Chart")

ax[1,1].hist(y)
ax[1,1].set_title("Histogram")

plt.suptitle("2x2 Subplot Layout")
plt.tight_layout()
plt.show()
```
